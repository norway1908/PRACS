```python
import pandas as pd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load MNIST CSV (Ensure to provide the correct file path)
file_path = "mnist_test.csv"  # Replace with your actual file path
mnist_data = pd.read_csv(file_path)

# Choose two random indices
random_index1 = np.random.randint(0, len(mnist_data))
random_index2 = np.random.randint(0, len(mnist_data))

# Extract the corresponding images and labels
label1 = mnist_data.iloc[random_index1, 0]  # First column is the label
random_image1 = mnist_data.iloc[random_index1, 1:].values.reshape(28, 28).astype(np.uint8)

label2 = mnist_data.iloc[random_index2, 0]  # First column is the label
random_image2 = mnist_data.iloc[random_index2, 1:].values.reshape(28, 28).astype(np.uint8)

# Convert to Pillow Images
img1 = Image.fromarray(random_image1)
img2 = Image.fromarray(random_image2)

# Display the Random Images
plt.figure(figsize=(7, 7))

# Image 1
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap="gray")
plt.title(f"Digit 1: {label1}")
plt.axis("off")

# Image 2
plt.subplot(1, 2, 2)
plt.imshow(img2, cmap="gray")
plt.title(f"Digit 2: {label2}")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
#EXP1 

# Convert to Grayscale (if needed, for demonstration purposes)
gray_img = img1.convert("L")

# Display Before and After
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap="gray")
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(gray_img, cmap="gray")
plt.title("Grayscale Image")
plt.show()
```

```python
# Crop the central 14x14 region
cropped_img = img1.crop((7, 7, 21, 21))  # (left, upper, right, lower)

# Display Before and After
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap="gray")
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(cropped_img, cmap="gray")
plt.title("Cropped Image")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Assuming 'random_image' is already loaded (from previous steps)
# Generate a dummy image (a constant matrix) for demonstration
dummy_img = np.full((28, 28), 50, dtype=np.uint8)  # Constant value of 50

# Prevent division by zero by replacing 0s with 1 in random_image2 for division
random_image2_safe = np.where(random_image2 == 0, 1, random_image2)

# Perform Arithmetic Operations
added_img = Image.fromarray(np.clip(random_image1 + random_image2, 0, 255))
subtracted_img = Image.fromarray(np.clip(random_image1 - random_image2, 0, 255))
multiplied_img = Image.fromarray(np.clip(random_image1 * random_image2, 0, 255))
divided_img = Image.fromarray(np.clip(random_image1 // random_image2_safe, 0, 255))

# Set the blending factor (you can adjust this value)
alpha = 0.7  # Adjust alpha between 0 and 1

# Perform Arithmetic Blending
blended_img_array = np.clip(alpha * random_image1 + (1 - alpha) * random_image2, 0, 255)
blended_img = Image.fromarray(blended_img_array.astype(np.uint8))


# Create subplots to display the images
plt.figure(figsize=(7, 7))

# Original Image
plt.subplot(2, 3, 1)
plt.imshow(random_image1, cmap="gray")
plt.title("Original Image 1")
plt.axis("off")

# Original Image
plt.subplot(2, 3, 2)
plt.imshow(random_image2, cmap="gray")
plt.title("Original Image 2")
plt.axis("off")

# Respective Operations next to the Original Image
# Added Image
plt.subplot(2, 3, 3)
plt.imshow(added_img, cmap="gray")
plt.title("Added Image")
plt.axis("off")

```
